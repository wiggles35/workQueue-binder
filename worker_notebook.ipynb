This is a program that launches your workers. Simply press shift+enter to run. 

Behind the scenes workers are being started that will run the tasks that the manager program is giving. The manager and the worker are connected by specifying the host and port of the master program. 

```python
import work_queue
workers = work_queue.Factory(batch_type='local', master_host_port='localhost:1024')
workers.cores = 1
with workers:
    workers.max_workers = 3
    
```
